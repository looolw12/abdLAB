```python
import time, datetime, csv, subprocess as sp, pandas as pd
```

```python
def speedtest(args = '--simple'):
    cmd = ['speedtest-cli'] + args.split()
    process = sp.Popen(cmd, stdout=sp.PIPE)
    output = process.communicate()[0].split()[1::3]
    output = [float(d) for d in output]
    return output + [datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")]
```

```python
def print_from_csv(filename):
    try:
        with open(filename + '.csv', 'r') as f:
            for row in csv.reader(f):
                print(row)
    except Exception as e:
        print(e)
```

```python
try:
    with open('speed.csv', 'a') as file:
        w = csv.writer(file)

        for i in range(100):
            current = speedtest()
            w.writerow(current)

            (p, d, u, t) = current
            print(f'#{i:3} Ping: {p:6.2f} Download {d:6.2f} Upload {u:6.2f} Datetime: {t}')
            time.sleep(0.5)

except KeyboardInterrupt:
    pass
except Exception as e:
    print(e)
```

```python
print_from_csv('speed')
```

```python
df = pd.read_csv('speed.csv', names = ['Ping (ms)', 'Download (Mbit/s)', 'Upload (Mbit/s)', 'Datetime'])
df.head()
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')
df['Date'] = df['Datetime'].apply(lambda x: x.date() if pd.notna(x) else None)
df['Time'] = df['Datetime'].apply(lambda x: x.time() if pd.notna(x) else None)
df.drop(columns=['Datetime'], inplace=True)
df.head()
```

```python
df.to_csv('speed_with_split_datetime.csv')
```

Что такое анонимная функция Python, как она работает
Функция которая не привязана к имени и может содерать только одно выражение.

Что возвращает функция speedtest? Какой код используется для просмотра результатов функции speedtest?
В оригинале результат "speedtest-cli --simple" разделяется пробелами на список строк и к этому списку добавляется Дата+время. В написаном мною коде в возвращаемом списке содержаться только значения ping, download, upload преобразованные в тип float и строка Дата+время. Для просмотра результата достаточно функции print.

Каким образом выполняется переименование столбцов датафрейма?
В оригинале используется функции rename с параметром inplace=True для прямого изменения, а не возвращения нового фрейма с изменеными именами. В моем коде столбцы сразу имеет правильное имя.

Для чего в лабораторной работе импортируется библиотека NumPy?
Ни в оригинале ни в моем коде библиотека не используется, но в описании библиотеки pandas написано что работа с данными строится поверх NumPy. Сама библиотека предоставляет оптимизированные типы и алгоритмы для работы с числами т.к. в Python типы данных имеет больший размер по сравнению с компилируемыми языками программирования.
